# How to deal with constraints ?

Although there are methods for dealing with constraints, one solution is to transform the problem so that it becomes unconstrained.

The basic idea is to add penalties to the cost function so that it increases significantly when the constraint is violated.


## Lagrange Mutliplier

### Equalities constraints

$\min f(x)$ with $h(x)=0$,

so that $x \in \mathbb R^n$

and $f:\mathbb{R}^n \to \mathbb{R}$

and $h:\mathbb{R}^n \to \mathbb{R}^m$ with $m<n$



**Exemple**

$$\min f(x)={x_1}^2+{x_2}^2$$
avec
$$h(x)={x_1}^2+2{x_2}^2-1=0$$


---
### Regular point


**Definition**
Let build the Jacobian relative to h

$$J(x)= \begin{bmatrix} \nabla h_1(x)^T \\ \vdots \\ \nabla h_m(x)^T \end{bmatrix}$$

$x^*$, so that  $h(x^*)=0$ is a **regulart point** of the constraints if
$$rank J(x^*)=m$$

---

**Surface**

The equality constraints $h(x)=0$ describe a surface in $\mathbb{R^n}$

$$S=\{x\in\mathbb{R}^n | h(x)=0\}$$
If the points of $S$ are regular, then $S$ is of dimension $(n-m)$

**Curve on a surface**

A curve ${C}$ on the surface $S$ is a set of points $\{ x(t)\in S \}$ continuously parameterised by $t\in\mathbb{R}$.

The curve $C$ going through a point $x^*$ if there exists $t^*$ such that $x(t^*)=x^*$.

The vector $\frac{dx}{dt|_{t=t^*}}$ is the tangent to the curve ${C}$ at the point $x^*$

**Tangent space**

The tangent space $T(x^*)$ at a point $x^*$ on the surface $S$ is defined as the set $T(x^*)=\{y\in\mathbb{R} | J(x^*)y=0\}$

If $(x^*)$ is regular, then $T(x^*)$ has dimension $(n-m)$.

Let ${C}$ be a curve on the surface $S$ passing through $x^*$. Then the tangent to the curve ${C}$ is $\frac{dx}{dt|_{t=t^*}}\in T(x^*)$

**Normal space**

The normal space $N(x^*)$ at a point $x^*$ on the surface $S$ is defined as the set $N(x^*)=\{y\in\mathbb{R}^n |y=J^T(x^*)z$ with $z \in \mathbb{R}^m\}$

If $x^*$ is regular, then $N(x^*)$ is of dimension $m$

**Property**

$$T(x^*) \perp N(x^*) \Leftrightarrow \forall u \in T(x^*), \forall v \in N(x^*), u^Tv=0$$

---
### First-order necessary condition
Let f(x) be minimum at $x^*$ under the constraint $h(x)=0$
If $x^*$ is a regular point, then
$$\exists \lambda^* \in \mathbb{R}^m | \nabla f(x^*) + \lambda ^*\nabla h^T(x^*)=0$$

---

**Demonstration**

Let $x(t)|h(x(t))=0 \forall t$ and therefore $\exists t^* | x(t^*)=x^*$,


If f(x(t)) is minimum at $t=t^*$, then $\frac{df(x(t))}{dt|_{t=t^*}}=0 \Leftrightarrow \nabla f^T(x^*) \frac{dx}{dt|_{t=t^*}}=0$

and therefore

$\nabla f^T(x^*) \in N(x^*)$

or $\nabla h^T(x^*)$ is a basis for the normal space $N(x^*)$,

and therefore

$\nabla f(x^*)$ is a linear combination of the columns of $\nabla h^T(x^*)$,

which is equivalent to writing

$$\exists \lambda^* \in \mathbb{R^m}|\nabla f(x^*) = - \lambda ^*\nabla h^T(x^*)$$


**Exemple**

$$\min f(x)={x_1}^2+{x_2}^2$$
avec
$$h(x)={x_1}^2+2{x_2}^2-1=0$$

### Necessary and sufficient conditions
Let $l(x,\lambda ) = f(x)+\lambda ^T h(x)$ be a Lagrange function
Let $L(x, \lambda)$ be the Hessian of $l$
We can write it as
$$L(x, \lambda) = \nabla (\nabla ^T f(x))+ \lambda_1 H_1(x)+ \dots + \lambda_m H_m(x)$$
or
$$L(x, \lambda) = H(x) + \sum_{i=1}^m \lambda_i H_i(x)$$

**Second order Necessary C**

Let $f(x)$ be the minimum at $x^*$ under the constraint h(x)=0

If $x^*$ is a regular point, then
$\exists \lambda^*  $ such that
1. $\nabla l(x^*,\lambda^*) = \nabla f(x^*) + \nabla h ^T (x^*)\lambda ^*=0$ and $h(x^*)=0$
2. $y^TL(x^*,\lambda^*) y \geq 0, \forall y \in T(x^*) \Leftrightarrow \forall y| y^T\nabla h ^T(x^*)=0 $

**Second-order Sufficient C**

If $\exists x^*, \exists \lambda^* $ such that
1. $\nabla f(x^*) + \nabla h ^T (x^*)\lambda ^*=0$ and $h(x^*)=0$
2. $y^TL(x^*,\lambda^*) y \geq 0, \forall y \neq 0 | y^T\nabla h ^T(x^*)=0 $

Then $x^*$ is a strict local minimum of f(x) with h(x)=0

## Karush Kuhn Tucker Condition

Now, let try to find the solutions of a problem with inequality constraints
\begin{align}
 \min f(x)& \\
 \mbox{avec } h(x)&=0 \\
 \mbox{et } c(x)&< 0
\end{align}


**Active/inactive**

$c_i(x)\leq 0 $ is **active** in x if $c_i(x)=0$.

It is **inactive** if $c_i(x)<0$

**Regular point**

$x^*$ satisfying the constraints $h(x^*)=0$ and $c_j(x^*)=0, \forall j \in C(x^*)$, j is the set of indexes of active constraints.

$x^*$ is a regular point of the active constraints if $\nabla h_i(x^*)$ and $\nabla c_j(x^*), \forall j \in C(x^*)$ are linearly independant.

### First Order Necessary Condition

Let $x^*$ minimise $f(x)$ under the constraints $h(x)=0$ and $c(x)\leq 0$

If $(x^*)$ is a regular point, then

$\exists \lambda^*, \mu^*$ so that

1. $\mu^* \geq 0$
2. $\nabla f(x^*)+\nabla h ^T(x^*) \lambda ^* + \nabla c^T(x^*)\mu^*$=0
3. ${\mu^*}^Tc(x^*)=0$

It implies that ${\mu^*}_i=0$ if the constraint $c_i(x^*)<0$, ie if the constraint is inactive.

##Necessary and sufficient conditions of the second order##

Let $L(x,\lambda,\mu)=H(x)+\sum_{i=0}^m \lambda_i H_i(x) + \sum_{j=1}^p \mu_j H_j(x) $

Where i are the indexes of the active constraints and j the indexes of the inactive constraints

---
**Second order CN**

Let $x^*$ that minimises f(x*) under the constraints $h(x)=0$ and $c(x)\leq 0$

If $x^*$ is a regular point, then

$\exists \lambda^*, \mu^* |$

1. $\mu^* \geq 0$
2. $\nabla f(x^*) + \nabla h^T(x^*)\lambda^* + \nabla c^T(x^*)\mu^*=0$
3. ${\mu^*}^T c(x^*)=0$
4. $y^T L(x^*,\lambda^*,\mu^*)y \geq 0, \forall y \neq 0 | \left\{ \begin{eqnarray} y^T\nabla h^T(x^*)& =0&\\ y^T\nabla {c_j}^T(x^*)& =0& \forall j \in C(x^*)\end{eqnarray} \right. $

---

**Second Order SC **


If $\exists \lambda^*, \mu^* |$

1. $\mu^* \geq 0$
2. $\nabla f(x^*) + \nabla h^T(x^*)\lambda^* + \nabla c^T(x^*)\mu^*=0$ and $h(x)=0$ and $c(x)\leq 0$
3. ${\mu^*}^T c(x^*)=0$
4. $y^T L(x^*,\lambda^*,\mu^*)y > 0, \forall y \neq 0 | \left\{ \begin{eqnarray} y^T\nabla h^T(x^*)& =0&\\ y^T\nabla {c_j}^T(x^*)& =0& \forall j \in C(x^*)\end{eqnarray} \right. $

then $x^*$ is a local minimum of $f(x)$ with $h(x)=0$ and $c(x)\leq 0$

# Resolution Method



## Projected gradient

This method is widely used when the constraints define a simple admissible space such as $\Omega = \mathbb{R^{n}_+}$ or $\Omega = \{x|m_i<x_i<M_i\}$.

The problem to be solved is in the form

$$\min f(x) $$

under $$D_1x=d_1$$ and $$D_2x\leq d_2$$

Suppose that we know $x_k$ as a point in the admissible space $\Omega$. At this point, we have determined which constraints are active: let $A_k$ be the matrix of these constraints. Note that this matrix $A_k$ necessarily contains $D_1$.
To determine $x_{k+1}$, we will move from  $x_k$ in the opposite direction of the projected gradient $g_k$, the latter being the solution of

$min \frac{1}{2} \| g-\nabla f(x_k)^2\|$ under $A_k g=0$

We then write the Lagrangian associated with this projection problem and we have

$$l(g,\lambda) = \frac{1}{2} \| g-\nabla f(x_k)^2\|+\lambda^T A_k g$$

The associated optimality conditions are

$g-\nabla f(x_k)+\lambda A_k^T=0$ and $A_kg=0$

If we multiply the first condition by $A_k$ on the left, we obtain

$$A_kg-A_k\nabla f(x_k)+A_kA_k^T\lambda=0$$

and using the second, we simplify:

$$-A_k\nabla f(x_k)+A_kA_k^T\lambda=0$$

If $A_k A_k^T$ is invertible, then

$$\lambda=(A_kA_k^T)^{-1}A_k\nabla f(x_k)$$

We substitute this value into the optimality condition and obtain:

$$g-\nabla f(x_k)+A_k^T\lambda=0\\
\Leftrightarrow g=\nabla f(x_k)+A_k^T(A_kA_k^T)^{-1}A_k\nabla f(x_k)$$

or

$$ g=(\mathbb{I}+A_k^T(A_kA_k^T)^{-1}A_k)\nabla f(x_k)$$

and  $(\mathbb{I}+A_k^T(A_kA_k^T)^{-1}A_k)=P$ is a projection of the gradient onto the active constraints. This is the matrix of the orthogonal projection onto the tangent subspace at $x_k$. The vectors $\nabla f(x_k)$ and $g$ are orthogonal.

We can then easily verify that if g is non-zero, then $-g$ is indeed a descent direction and all that remains is to move in that direction.

We must find $\alpha$ such that $x_{k+1}=x_{k}-\alpha g$ is admissible.

We then encounter two cases:

either

1. $\alpha$ is the largest possible value that preserves admissibility, in which case, at the next rank, $A_{k+1}$ must be updated by adding the new active constraints
2. or $\alpha$ is not the largest possible value, in which case it must be verified that the new active constraints are the same as in $A_k$ and removed if necessary.






##  Penalty method

Let's consider the problem

$$min f(x)$$ with $$c(x)\leq 0$$

We transform this problem into an unconstrained optimisation problem:
$$\min g(x, \alpha_k)=f(x)+\alpha_k \sum_{i=1}^p P_i(c_i(x))$$

With
1. $P(c(x))$ continuous
2. $P(c(x)) \geq 0$, $\forall x$
3. $\alpha_k \geq 0$

### Internal penalty method

$c(x)\leq 0$

The penalties are close to zero far from the constraints and tend towards infinity at the limits of the constraints.

For example, we choose the functions P:
1. $P(x) = \frac{-1}{c(x)}$
2. $P(x) = \frac{1}{c(x)^2}$
3. $P(x) = ln(1 - \frac{1}{c(x)})$

**Example**
Let $f(x)=2x$ and $c(x)=1-x\leq0$

If we choose $P(x)= \frac{-1}{c(x)}$,

The problem becomes: $min g(x, \alpha_k)=2x+\alpha_k\frac{1}{x-1}$

### External penalty method
$c(x)\leq 0$

Penalties are zero in the admissible space and positive when the constraints are no longer satisfied.

For example
1. $P(c(x))= \left\{\begin{eqnarray}c(x)^q \text{ if } c(x)>0\\ 0 \text{ if } c(x)\leq 0\end{eqnarray}\right. = max(c(x),0)^q$
2. $P(c(x))= \left\{\begin{eqnarray}a^{c(x)}-1 \text{ if } c(x)>0\\ 0 \text{ if } c(x)\leq 0\end{eqnarray}\right.$

**Example**

Let $f(x)=2x$ and $c(x)=1-x\leq0$

If $P(x)=max(1-x,0)^2$

The problem boils down to minimising
$$\min g(x,\alpha_k) = 2x +\alpha_k max(1-x,0)^2$$


```python
x=np.linspace(0.5,2,100)
plt.plot(x,2*x)
plt.plot(np.array([1,1]),np.array([0,5]),'r')
y = (1-x)*((1-x)>0)
alpha = 2
plt.plot(x,2*x+alpha*(y)**2,'g')
alpha = 4
plt.plot(x,2*x+alpha*(y)**2,'c')
alpha = 6
plt.plot(x,2*x+alpha*(y)**2,'c')
```
